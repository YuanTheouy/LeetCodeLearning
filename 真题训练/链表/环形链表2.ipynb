当然，这个问题还有进阶版，也是力扣第 142 题「环形链表 II」：如果链表中含有环，如何计算这个环的起点？

举个例子，环的起点是指下面这幅图中的节点 2：


这里先直接看一下寻找环起点的解法代码：

```python
class Solution:
    def detectCycle(self,head):
        fast,slow = head,head
        while fast and fast.next:
            fast = fast.next.next
            slow = slow.next

            if fast == slow:
                break
        
        if not fast or not fast.next:
            return None
        
        slow = head
        while slow!=fast:
            slow = slow.next
            fast = fast.next
        
        return slow
```

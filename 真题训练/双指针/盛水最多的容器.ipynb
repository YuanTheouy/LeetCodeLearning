```python
def maxArea(height):
    n = len(height)
    left,right = 0, n-1
    res = 0
    while left<right:
        min_height = min(height[left],height[right])
        cur_area = min_height * (right-left)# æ³¨æ„ï¼Œè¿™ä¸ªæ˜¯è–„å£ï¼Œæ‰€ä»¥ä¸èƒ½-1
        res = max(res,cur_area)

        if height[left]<height[right]:
            left+=1
        else:
            right-=1

    return res
```

给你一个整数数组 nums 和一个整数 k ，请你返回其中出现频率前 k 高的元素。你可以按 任意顺序 返回答案。

 

示例 1：

输入：nums = [1,1,1,2,2,3], k = 2

输出：[1,2]

示例 2：

输入：nums = [1], k = 1

输出：[1]

示例 3：

输入：nums = [1,2,1,2,1,2,3,1,3,2], k = 2

输出：[1,2]



```python
import heapq

class Solution:
    def topKFrequent(self, nums, k):
        number = {}
        pq = []
        res = []
        for i,num in enumerate(nums):
            number[num] = number.get(num,0) + 1
            heapq.heappush(pq,(number[num],i,num))

            if len(pq)>k:
                heapq.heappop(pq)
            
        for _,_,num in pq:
            res.append(num)

        return num
```

```python
import heapq
class Solution:
    def topKFrequent(self,nums,k):
        count_map = {}
        for num in nums:
            count_map[num] = count_map.get(num,0)+1
        
        pq = []
        res = []
        for num, count in count_map.items():
            heapq.heappush(pq,(count,num))
        
            if len(pq)>k:
                heapq.heappop(pq)
            
        for _,x in pq:
            res.append(x)

        return res
```

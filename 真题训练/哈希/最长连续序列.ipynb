```python
class solution:
    def longestConseculative(self,nums):
        set_nums = set(nums)

        res = 0

        for num in set_nums:
            if num-1 in set_nums:
                continue

            cur_num = num
            cur_len = 1

            while cur_num+1 in set_nums:
                cur_num+=1
                cur_len+=1

            res = max(res,cur_len)

        return res

# --- æµ‹è¯•éƒ¨åˆ† ---
if __name__ == "__main__":
    # æµ‹è¯•ç”¨ä¾‹ 1
    nums = [100, 4, 200, 1, 3, 2]
    sol = solution()
    print(f"è¾“å…¥: {nums}")
    print(f"è¾“å‡º: {sol.longestConseculative(nums)}") # é¢„æœŸ 4 ([1,2,3,4])
    
    # æµ‹è¯•ç”¨ä¾‹ 2
    nums2 = [0, 3, 7, 2, 5, 8, 4, 6, 0, 1]
    print(f"\nè¾“å…¥: {nums2}")
    print(f"è¾“å‡º: {sol.longestConseculative(nums2)}") # é¢„æœŸ 9
    
    # --- éªŒè¯ä½ çš„ç–‘é—®ï¼šSet åˆ°åº•æ˜¯ä¸æ˜¯æœ‰åºçš„ï¼Ÿ---
    print("\n--- Set é¡ºåºå®éªŒ ---")
    demo_list = [3, 1, 2]
    print(f"List: {demo_list} -> Set: {set(demo_list)}") 
    # è¿™é‡Œçš„è¾“å‡ºå¾ˆå¯èƒ½è®©ä½ è§‰å¾—æ˜¯æœ‰åºçš„ï¼Œæ¯”å¦‚ {1, 2, 3}


```

# 链表

```python
class ListNode:
    def __init__(self,x):
        self.val = x
        self.next = None
```

```python
# 双链表节点
class Node:
    def __init__(self,prev,element,next):
        self.val = element
        self.next = next
        self.prev = prev
```

```python

class ListNode:
    def __init__(self,x):
        self.val = x
        self.next = None

def creartLinkedList(arr: 'list[int]') -> 'ListNode':
    if arr is None or len(arr)==0:
        return None
    head = ListNode(arr[0])
    cur = head
    for i in range(1,len(arr)):
        cur.next = ListNode(arr[i])
        cur = cur.next

    return head



```

```python
# 查改

head = creartLinkedList([1,2,3,4,5])

p = head

while p is not None:
    print(p.val)
    p = p.next
```

```python
# 增
## 头部增加节点

head = creartLinkedList([1,2,3,4,5])
newNode = ListNode(0)
newNode.next = head
head = newNode
```

```python
## 单链尾部增加新元素
head = creartLinkedList([1,2,3,4,5])

p = head

while p.next is not None:
    p = p.next

p.next = ListNode(6)
```

```python
## 单链中部增加元素

head = creartLinkedList([1,2,3,4,5])

p = head
# 第三个节点后插入

for _ in range(2):
    p = p.next

newNode = ListNode(66)
newNode.next = p.next
p.next = newNode
```

```python
# 删
## 删除第四个节点
head = creartLinkedList([1,2,3,4,5])

p = head

for i in range(2):
    p = p.next

p.next = p.next.next
```

```python
# 删除尾部元素
head = creartLinkedList([1,2,3,4,5])

p = head


while p.next.next is not None:
    p = p.next

p.next = None
```

```python
# 删除单链表头部元素

head = creartLinkedList([1,2,3,4,5])

head = head.next
```

```python
# 双链表基本操作

class DoublyListNode:
    def __init__(self,x):
        self.val = x
        self.prev = None
        self.next = None

def createDoublyLinkedList(arr: list[int]):
    if not arr:
        return None
    
    head = DoublyListNode(arr[0])

    cur = head

    for val in arr[1:]:
        new_node = DoublyListNode(val)
        new_node.prev = cur
        cur.next =new_node
        cur = new_node

    return head
```

```python
# 查改

# 双链表 可以从头或尾开始，遍历

head = createDoublyLinkedList([1,2,3,4,5])
tail = None

# 从头向尾找
p = head
while p:
    print(p.val)
    tail = p
    p = p.next

# 从尾向前找
p = tail
while p:
    print(p.val)
    p = p.prev
```

```python
# 增
# 头部插入元素
head =createDoublyLinkedList([1,2,3,4,5])

new_head = DoublyListNode(0)

new_head.next = head
head.prev = new_head
head = new_head
```

```python
# 尾部增
head = createDoublyLinkedList([1,2,3,4,5])
tail = head
while tail.next is not None:
    tail = tail.next

new_tail = DoublyListNode(6)
new_tail.prev = tail
tail.next = new_tail

tail = new_tail
```

```python
# 中间插入元素
head = createDoublyLinkedList([1,2,3,4,5])

p = head
for _ in range(2):
    p = p.next

newNode = DoublyListNode(66) #插入到第四个节点位置

newNode.next = p.next
p.next.prev = newNode

p.next = newNode
newNode.prev = p

for _ in range(6):
    print(head.val)
    head = head.next
```

```python
# 删
# 删除第4个节点
head =createDoublyLinkedList([1,2,3,4,5])

p = head
for i in range(2):
    p = p.next

toDelete = p.next

p.next = toDelete.next
toDelete.next.prev = p

toDelete.next =None
toDelete.prev = None



```

# 层序遍历 BFS
## 一层层的遍历二叉树

```python
from collections import deque

def level0derTraverse(root):
    if root is None:
        return
    q = deque()
    q.append(root)

    while q:
        cur = q.popleft()
        print(cur.val)

        if cur.left is not None:
            q.append(cur.left)

        if cur.right is not None:
            q.append(cur.right)
```

```python
from collections import deque

def traverse(root):
    if root is None:
        return
    
    q = deque()
    q.append(root)

    depth = 1

    while q:
        sz = len(q)
        for i in range(sz):
            cur = q.popleft()
            print(f"depth={depth},val={cur.val}")

            if cur.left is not None:
                q.append(cur.left)
            if cur.right is not None:
                q.append(cur.right)

        depth += 1
```

```python
from collections import deque

def traverse(root):
    if root is None:
        return
    q = deque()
    q.append(root)
    depth = 1
    while q:
        sz = len(q)
        for i in range(sz):
            cur = q.leftpop()

            print(f"depth:{depth},val:{cur.val}")
            if cur.left is not None:
                q.append(cur.left)
            if cur.right is not None:
                q.append(cur.right)

        depth += 1
```

```python
class MyPriorityQueue:
    # 堆顶插入一个数据
    def push(self,x):
        pass
    # 返回堆顶
    def peek(self):
        pass
    # 删除堆顶
    def pop(self):
        pass
    # 返回堆中元素个数,O(1)
    def size(self):
        pass
```

```python
def heapSort(arr):
    res = [0] * len(arr)
    pq = MyPriorityQueue()
    for x in arr:
        pq.push(x)
    for i in range(len(arr)):
        res[i] = pq.pop()
    return res
```

```python
import heapq

def heapSort(arr):
    pq = []
    for x in arr:
        heapq.heappush(pq,x)
    res = []

    while pq:
        res.append(heapq.heappop(pq))

    return res

# 测试
arr = [3, 1, 4, 1, 5, 9, 2, 6]
print(heapSort(arr))  # 输出: [1, 1, 2, 3, 4, 5, 6, 9]
```

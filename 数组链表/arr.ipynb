# 数组 静态和动态简单用法

```python
arr = [0] * 10
arr
```

```python
arr[0] = 2
arr[1] = 1
arr
```

```python
# 增
## append
arr = [0] * 10
for i in range(4):
    arr[i]=i

arr[4] = 4
arr[5] = 5

arr.append(6)
arr
```

```python
## insert

arr = [0]*10
for i in range(4):
    arr[i]=i

# for i in range(4,2,-1):
    # arr[i] = arr[i-1]

# arr[2] = 666



for i in range(4,2,-1):
    arr[i]=arr[i-1]

arr
```

```python
# 扩容增
arr = [i for i in range(10)]
newArr = [0]*20

for i in range(10):
    newArr[i]=arr[i]

newArr[10] =10
newArr
```

```python
# 删
## 删除末尾元素
arr = [0]*10
for i in range(5):
    arr[i] = i

arr[4] = -1 # 用负一替代

arr
```

```python
# 删中间,删第二个元素，i=1
arr = [0]*10
for i in range(5):
    arr[i] = i

for i in range(1,4):
    arr[i] = arr[i+1]

arr[4] = -1


for i in range(1,4):
    arr[i] = arr[i+1]

```

增：
    append：O(1)
    insert: O(n)
删：
    删尾：O(1)
    删中间: O(N)
查：O(1)
改：O(1)
如果反过来，比方说给你一个值，让你去找这个值在数组中对应的索引，那你只能遍历整个数组去寻找对吧，这个复杂度就是 O(N) 了

```python
# 动态数组

arr = []
for i in range(10):
    arr.append(i)

arr.insert(2,666)
arr.insert(0,-1)

# 删除尾部元素
arr.pop()
#删除中间元素(删除索引为2的元素)
arr.pop(2)

# 根据序号改 O(1)
arr[0]=100

# 查序号， O(N)
index = arr.index(666)

arr
```
